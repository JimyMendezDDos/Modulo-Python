```python
#crear una calculadora simple
try:
    signo = input("Ingrese un signo de operación (+, -, *, /): ")
    num1 = float(input("Ingrese el primer número: "))
    num2 = float(input("Ingrese el segundo número: "))

    if signo == "+":
        resultado = num1 + num2
        print("Tu resultado es: "+ str(resultado))
    elif signo == "-":
        resultado = num1 - num2
        print("Tu resultado es: "+ str(resultado))
    elif signo == "*":
        resultado = num1 * num2
        print("Tu resultado es: "+ str(resultado))
    elif signo == "/":
        if num2 != 0:
            resultado = num1 / num2
            print("Tu resultado es: "+ str(resultado))
        else:
            print("Error: División por cero no permitida.")
    else:
        print("Error: Signo de operación no válido.")
        
except ValueError:  #si atrapa un error de valor, seguira ejecutandose el programa
    print("Ingrese un dato valido.")
    
    #Prueba de comentario adicional
```
